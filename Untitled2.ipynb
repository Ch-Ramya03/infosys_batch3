```python
import os
import zipfile
import warnings
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

warnings.filterwarnings("ignore")
zip_path = "city_day.csv (1).zip"  # <-- put your path here

with zipfile.ZipFile(zip_path, "r") as zf:
    csv_name = [n for n in zf.namelist() if n.endswith(".csv")][0]
    with zf.open(csv_name) as f:
        df = pd.read_csv(f)

print(f"Loaded {csv_name} with shape {df.shape}")
print(df.head())
```

```python
print("Shape:", df.shape)
print("\nColumns:\n", df.columns.tolist())
print("\nMissing values per column:\n", df.isnull().sum() )
```

```python
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
# include pollutant-like object columns
possible_pollutants = ["pm2.5","pm25","pm10","no2","so2","co","o3","aqi"]
for c in df.columns:
    if any(p in c.lower() for p in possible_pollutants):
        df[c] = pd.to_numeric(df[c], errors="coerce")
        if c not in numeric_cols:
            numeric_cols.append(c)

# Remove 'AQI_Bucket' from numeric_cols if it exists
if 'AQI_Bucket' in numeric_cols:
    numeric_cols.remove('AQI_Bucket')

print("Numeric pollutant columns:", numeric_cols)
```

```python
df[numeric_cols].isna().sum()
```

```python
summary = df[numeric_cols].describe().T
summary["skew"] = df[numeric_cols].skew()
summary["kurt"] = df[numeric_cols].kurtosis()
print("\nSummary Statistics:\n", summary)
```

```python
for col in numeric_cols[:6]:
    plt.figure(figsize=(10,3))
    plt.plot(df.index, df[col])
    plt.title(col)
    plt.xlabel("Date")
    plt.tight_layout()
    plt.show()
```

```python
corr = df[numeric_cols].corr()
plt.figure(figsize=(8,6))
plt.imshow(corr, cmap="coolwarm", aspect="auto")
plt.xticks(range(len(corr.columns)), corr.columns, rotation=45, ha="right")
plt.yticks(range(len(corr.index)), corr.index)
plt.colorbar(label="Correlation")
plt.title("Pollutant Correlation Matrix")
plt.tight_layout()
plt.show()
```

```python
monthly = df[numeric_cols].resample("M").median()
plt.figure(figsize=(10,4))
for col in numeric_cols[:6]:
    plt.plot(monthly.index, monthly[col], label=col)
plt.legend()
plt.title("Monthly Median Trends")
plt.tight_layout()
plt.show()
```

```python
weekly = df[numeric_cols].resample("W").mean()
monthly_mean = df[numeric_cols].resample("M").mean()
```

```python
def make_forecast_features(df_ts, target_col, lags=[1,2,3,7], rolling_windows=[3,7,14]):
    X = pd.DataFrame(index=df_ts.index)
    X[target_col] = df_ts[target_col]
    for lag in lags:
        X[f"{target_col}_lag{lag}"] = df_ts[target_col].shift(lag)
    for w in rolling_windows:
        X[f"{target_col}_roll_mean_{w}"] = df_ts[target_col].rolling(w).mean().shift(1)
        X[f"{target_col}_roll_std_{w}"] = df_ts[target_col].rolling(w).std().shift(1)
    # add lag features for other pollutants
    for col in [c for c in numeric_cols if c != target_col]:
        X[f"{col}_lag1"] = df_ts[col].shift(1)
        X[f"{col}_lag7"] = df_ts[col].shift(7)
    # date features
    X["dayofweek"] = X.index.dayofweek
    X["month"] = X.index.month
    X["is_weekend"] = X.index.dayofweek.isin([5,6]).astype(int)
    X = X.dropna()
    return X

target = "PM2.5" if "PM2.5" in numeric_cols else numeric_cols[0]
features = make_forecast_features(df, target)
print("Feature matrix shape:", features.shape)

# Train/test split (last 30 days)
test_start = features.index.max() - pd.Timedelta(days=30)
train = features[features.index <= test_start]
test = features[features.index > test_start]

# ---------------------------------
# 8. Save processed outputs
# ---------------------------------
os.makedirs("processed", exist_ok=True)
df.to_csv("processed/clean_city_day_interpolated.csv")
weekly.to_csv("processed/weekly_resampled.csv")
monthly_mean.to_csv("processed/monthly_resampled.csv")
features.to_csv("processed/forecast_features.csv")
train.to_csv("processed/forecast_train.csv")
test.to_csv("processed/forecast_test.csv")

print("\nSaved processed files in './processed/' folder.")
print("Done ✅")
```
