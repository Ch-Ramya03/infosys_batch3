```python
import os
import zipfile
import warnings
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

warnings.filterwarnings("ignore")
zip_path = "city_day.csv.zip"  # <-- put your path here

with zipfile.ZipFile(zip_path, "r") as zf:
    csv_name = [n for n in zf.namelist() if n.endswith(".csv")][0]
    with zf.open(csv_name) as f:
        df = pd.read_csv(f)

print(f"Loaded {csv_name} with shape {df.shape}")
print(df.head())
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df = df.dropna(subset=['PM2.5', 'PM10', 'NO2', 'AQI'])
city_data = df[df['City'] == 'Delhi'].copy()
city_data['AQI'] = city_data['AQI'].astype(float)
```

```python
plt.figure(figsize=(10,5))
plt.plot(city_data['Date'], city_data['PM2.5'], color='blue')
plt.title('Line Plot: PM2.5 Trend in Delhi')
plt.xlabel('Date')
plt.ylabel('PM2.5')
plt.show()
```

```python
plt.figure(figsize=(7,5))
plt.scatter(city_data['PM2.5'], city_data['AQI'], alpha=0.6, color='green')
plt.title('Scatter Plot: PM2.5 vs AQI')
plt.xlabel('PM2.5')
plt.ylabel('AQI')
plt.show()
```
